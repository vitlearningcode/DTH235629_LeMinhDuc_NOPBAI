```python
import tkinter as tk
from tkinter import messagebox

# Giả định mật khẩu cũ mặc định
DEFAULT_OLD_PASSWORD = "password123" 

def check_password():
    """Hàm xử lý logic kiểm tra mật khẩu khi nhấn nút OK."""
    # Lấy giá trị từ các trường nhập
    old_pass = entry_old.get()
    new_pass_1 = entry_new.get()
    new_pass_2 = entry_new_again.get()

    # --- Kiểm tra 1: Mật khẩu mới có giống mật khẩu cũ không? ---
    # Trong môi trường thực tế, old_pass sẽ được so sánh với mật khẩu lưu trong CSDL/biến
    if new_pass_1 == DEFAULT_OLD_PASSWORD:
        # Hiển thị thông báo nếu mật khẩu mới trùng mật khẩu cũ
        messagebox.showerror("Lỗi đổi mật khẩu", "Bạn không được dùng mật khẩu cũ")
        # Xóa các trường nhập mật khẩu mới để người dùng nhập lại
        entry_new.delete(0, tk.END)
        entry_new_again.delete(0, tk.END)
        return

    # --- Kiểm tra 2: Hai lần nhập mật khẩu mới có giống nhau không? ---
    if new_pass_1 != new_pass_2:
        # Hiển thị thông báo nếu hai lần nhập không giống nhau
        messagebox.showerror("Lỗi nhập liệu", "Bạn phải nhập lại mật khẩu mới giống nhau")
        # Xóa các trường nhập mật khẩu mới để người dùng nhập lại
        entry_new.delete(0, tk.END)
        entry_new_again.delete(0, tk.END)
        return

    # --- Trường hợp thành công ---
    if not new_pass_1 or not new_pass_2:
        messagebox.showwarning("Cảnh báo", "Vui lòng nhập đầy đủ mật khẩu mới.")
        return

    # Giả lập việc đổi mật khẩu thành công
    messagebox.showinfo("Thành công", "Đổi mật khẩu thành công!")
    root.destroy() # Đóng cửa sổ sau khi thành công

def cancel_action():
    """Hàm xử lý khi nhấn nút Cancel."""
    root.destroy()

# --- Thiết lập cửa sổ chính ---
root = tk.Tk()
root.title("Enter New Password")
root.geometry("350x200") # Thiết lập kích thước cửa sổ
root.resizable(False, False)

# --- Thiết lập Label và Entry cho "Old Password" ---
label_old = tk.Label(root, text="Old Password:", font=("Arial", 10))
label_old.grid(row=0, column=0, padx=10, pady=5, sticky="w")
entry_old = tk.Entry(root, width=30, show="*", font=("Arial", 10))
entry_old.grid(row=0, column=1, padx=10, pady=5)

# --- Thiết lập Label và Entry cho "New Password" ---
label_new = tk.Label(root, text="New Password:", font=("Arial", 10))
label_new.grid(row=1, column=0, padx=10, pady=5, sticky="w")
entry_new = tk.Entry(root, width=30, show="*", font=("Arial", 10))
entry_new.grid(row=1, column=1, padx=10, pady=5)

# --- Thiết lập Label và Entry cho "Enter New Password Again" ---
label_new_again = tk.Label(root, text="Enter New Password Again:", font=("Arial", 10))
label_new_again.grid(row=2, column=0, padx=10, pady=5, sticky="w")
entry_new_again = tk.Entry(root, width=30, show="*", font=("Arial", 10))
entry_new_again.grid(row=2, column=1, padx=10, pady=5)

# --- Thêm khoảng trống (padding) để căn chỉnh ---
tk.Label(root, text="").grid(row=3, column=0, columnspan=2, pady=5) 

# --- Thiết lập Nút OK và Cancel ---
button_ok = tk.Button(root, text="OK", command=check_password, width=10, font=("Arial", 10, "bold"))
button_ok.grid(row=4, column=0, padx=20, pady=10, sticky="e")

button_cancel = tk.Button(root, text="Cancel", command=cancel_action, width=10, font=("Arial", 10))
button_cancel.grid(row=4, column=1, padx=20, pady=10, sticky="w")

# Chạy vòng lặp sự kiện chính của Tkinter
root.mainloop()
```
