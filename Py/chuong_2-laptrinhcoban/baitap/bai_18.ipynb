```python
n = int(input("Nhập n: "))

# a) Đếm số ước nguyên tố khác nhau
def is_prime(x):
    if x < 2: return False
    for i in range(2, int(x**0.5) + 1):
        if x % i == 0: return False
    return True

dem = 0
for i in range(1, n+1):
    if n % i == 0 and is_prime(i):
        dem += 1

# b) Tổng các ước thực sự của n
tong_uoc = 0
for i in range(1, n):
    if n % i == 0:
        tong_uoc += i

# c) Tổng bình phương 1^2 + 2^2 + ... + n^2
tong_bp = 0
for i in range(1, n+1):
    tong_bp += i*i

print("a) Số ước nguyên tố khác nhau =", dem)
print("b) Tổng các ước thực sự =", tong_uoc)
print("c) Tổng bình phương =", tong_bp)
```

```python
n = int(input("Nhập n: "))

def is_prime(x):
    if x < 2: return False
    for i in range(2, int(x**0.5)+1):
        if x % i == 0: return False
    return True

# a) số ước nguyên tố khác nhau
dem = len({i for i in range(2, n+1) if n % i == 0 and is_prime(i)})

# b) tổng ước thực sự
tong_uoc = sum(i for i in range(1, n) if n % i == 0)

# c) tổng bình phương
tong_bp = sum(i*i for i in range(1, n+1))

print(f"a) {dem}\nb) {tong_uoc}\nc) {tong_bp}")
```

```python
import math

n = int(input("Nhập n: "))
m = n
prime_factors = set()

# Phân tích thừa số nguyên tố
for i in range(2, int(math.sqrt(n)) + 1):
    if m % i == 0:
        prime_factors.add(i)
        while m % i == 0:
            m //= i
if m > 1:
    prime_factors.add(m)

# a) Số ước nguyên tố khác nhau
dem = len(prime_factors)

# b) Tổng ước số thực sự
# Tính tổng tất cả ước bằng phân tích thừa số
m = n
sum_uoc = 1
for p in prime_factors:
    exp = 0
    while m % p == 0:
        exp += 1
        m //= p
    sum_uoc *= (p**(exp+1) - 1)//(p-1)
tong_uoc = sum_uoc - n   # bỏ n đi

# c) Tổng bình phương
tong_bp = n*(n+1)*(2*n+1)//6

print(f"a) {dem}\nb) {tong_uoc}\nc) {tong_bp}")
```
