```python
M = [3, 6, 7, 8, 11, 17, 2, 90, 2, 5, 4, 5, 8]

# kiểm tra số nguyên tố
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# số lẻ
odd = [x for x in M if x % 2 == 1]
print("Số lẻ:", odd, "→ có", len(odd), "số lẻ")

# số chẵn
even = [x for x in M if x % 2 == 0]
print("Số chẵn:", even, "→ có", len(even), "số chẵn")

# số nguyên tố
prime = [x for x in M if is_prime(x)]
print("Nguyên tố:", prime)

# không nguyên tố
non_prime = [x for x in M if not is_prime(x)]
print("Không nguyên tố:", non_prime)
```

```python
M = [3, 6, 7, 8, 11, 17, 2, 90, 2, 5, 4, 5, 8]

def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0: return False
    return True

def tach_day(M):
    odd, even, prime, non_prime = [], [], [], []
    for x in M:
        if x % 2: odd.append(x)
        else: even.append(x)
        if is_prime(x): prime.append(x)
        else: non_prime.append(x)
    return odd, even, prime, non_prime

odd, even, prime, non_prime = tach_day(M)

print("Dãy lẻ:", odd, f"(tổng {len(odd)})")
print("Dãy chẵn:", even, f"(tổng {len(even)})")
print("Dãy nguyên tố:", prime)
print("Dãy không nguyên tố:", non_prime)
```

```python
M = [3, 6, 7, 8, 11, 17, 2, 90, 2, 5, 4, 5, 8]

is_prime = lambda n: n > 1 and all(n % i for i in range(2, int(n**0.5)+1))

groups = {
    "odd": [x for x in M if x % 2],
    "even": [x for x in M if x % 2 == 0],
    "prime": [x for x in M if is_prime(x)],
    "non_prime": [x for x in M if not is_prime(x)],
}

print("Số lẻ:", groups["odd"], f"(tổng {len(groups['odd'])})")
print("Số chẵn:", groups["even"], f"(tổng {len(groups['even'])})")
print("Nguyên tố:", groups["prime"])
print("Không nguyên tố:", groups["non_prime"])
```
