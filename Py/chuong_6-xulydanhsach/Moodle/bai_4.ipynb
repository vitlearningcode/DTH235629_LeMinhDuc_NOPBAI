```python
#HHEHEHE
lst = [3, 0, 1, 5, 2]
x = 2

print("lst[0] =", lst[0])
print("lst[3] =", lst[3])
print("lst[x] =", lst[x])
print("lst[-x] =", lst[-x])
print("lst[x + 1] =", lst[x + 1])
print("lst[x] + 1 =", lst[x] + 1)
print("lst[lst[x]] =", lst[lst[x]])
print("lst[lst[lst[x]]]] =", lst[lst[lst[x]]])
```

```python
lst = [3, 0, 1, 5, 2]
x = 2

# a
a = lst[0]         # chỉ số 0 → 3
# b
b = lst[3]         # chỉ số 3 → 5
# c
c = lst[x]         # lst[2] → 1
# d
d = lst[-x]        # lst[-2] → lst[3] → 5
# e
e = lst[x + 1]     # lst[3] → 5
# f
f = lst[x] + 1     # lst[2] + 1 → 1 + 1 → 2
# g
g = lst[lst[x]]    # lst[lst[2]] → lst[1] → 0
# h
h = lst[lst[lst[x]]] # lst[lst[lst[2]]] → lst[lst[1]] → lst[0] → 3

print(a,b,c,d,e,f,g,h)
```

```python
lst = [3, 0, 1, 5, 2]
x = 2

results = {
    'lst[0]': lst[0],
    'lst[3]': lst[3],
    'lst[x]': lst[x],
    'lst[-x]': lst[-x],
    'lst[x + 1]': lst[x + 1],
    'lst[x] + 1': lst[x] + 1,
    'lst[lst[x]]': lst[lst[x]],
    'lst[lst[lst[x]]]': lst[lst[lst[x]]]
}

for expr, val in results.items():
    print(f"{expr} = {val}")
```
