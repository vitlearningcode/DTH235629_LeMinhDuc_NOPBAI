### **Các hàm**

```python
def sum1(n):
    s = 0
    while n > 0:
        s += 1
        n -= 1
    return s
```

* `sum1(n)` **không dùng global**, chỉ đếm số bước từ `n` về 0 → luôn trả về `n`.

---

```python
def sum2():
    global val
    s = 0
    while val > 0:
        s += 1
        val -= 1
    return s
```

* `sum2()` dùng **biến global val**
* Nó **giảm val về 0**
* Trả về số lần lặp = giá trị ban đầu của `val`

---

```python
def sum3():
    s = 0
    for i in range(val, 0, -1):
        s += 1
    return s
```

* `sum3()` dùng **giá trị hiện tại của val**
* Không thay đổi `val`

---

### **Trường hợp 1**

```python
def main():
    global val
    val = 5
    print(sum1(5))  # 5
    print(sum2())    # val = 5 → giảm về 0 → in 5
    print(sum3())    # val = 0 → vòng for không chạy → in 0
main()
```

**Kết quả:**

```
5
5
0
```

---

### **Trường hợp 2**

```python
def main():
    global val
    val = 5
    print(sum1(5))  # 5
    print(sum3())    # val = 5 → vòng chạy 5 lần → 5
    print(sum2())    # val = 5 → giảm về 0 → 5
main()
```

**Kết quả:**

```
5
5
5
```

✅ Lưu ý: `sum3()` chạy **trước sum2()**, nên val chưa bị thay đổi.

### **Trường hợp 3**

```python
def main():
    global val
    val = 5
    print(sum2())    # val = 5 → giảm về 0 → 5
    print(sum1(5))   # 5
    print(sum3())    # val = 0 → vòng for không chạy → 0
main()
```
