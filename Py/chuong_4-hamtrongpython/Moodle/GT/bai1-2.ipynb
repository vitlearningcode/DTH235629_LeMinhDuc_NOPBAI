```python
#1 - cb
def gen_even_basic(n):
    evens = []
    for i in range (n) :
        evens.append(2 * i)
    return evens

print (gen_even_basic(5))
```

```python
#1- nc
def gen_even_advanced (n):
    return [2 * i for i in range (n)]

print (gen_even_advanced (6) )
```

```python
#1b -cb
def gen_square_basic (n):
    square = []
    for i in range (n):
        square.append (i*i)
    return square

print (gen_square_basic (7))
```

```python
def gen_square_optimal (n):
    return [i * i for i in range (n)]

print (gen_square_optimal (7))
```
